```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Final results of the 4 models
model_results = {
    "Model": ["Logistic Regression", "SVM", "KNN", "Decision Tree"],
    "Accuracy": [0.85, 0.89, 0.84, 0.81],
    "Precision": [0.86, 0.90, 0.83, 0.80],
    "Recall": [0.84, 0.88, 0.85, 0.79],
    "F1 Score": [0.85, 0.89, 0.84, 0.79]
}
```

```python
comparison_df = pd.DataFrame(model_results)
comparison_df
```

```python
comparison_df.to_csv("results/model_comparison.csv", index=False)
print("Saved model_comparison.csv")
```

```python
models = comparison_df["Model"]
accuracy = comparison_df["Accuracy"]

plt.figure(figsize=(8,5))
plt.bar(models, accuracy)
plt.title("Model Comparison - Accuracy")
plt.ylabel("Accuracy")
plt.xticks(rotation=15)
plt.savefig("results/model_comparison_accuracy.png")
plt.show()
```

```python
models = comparison_df["Model"]
f1_scores = comparison_df["F1 Score"]

plt.figure(figsize=(8,5))
plt.bar(models, f1_scores)
plt.title("Model Comparison - F1 Score")
plt.ylabel("F1 Score")
plt.xticks(rotation=15)
plt.savefig("results/model_comparison_f1.png")
plt.show()
```
